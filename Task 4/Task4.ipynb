# Plotting using seaborn:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = r"E:\Data Visualisation Python Project\birthYearly.csv"
df = pd.read_csv(path)
df
```

We will be using heatmaps for this task:

Heatmaps are popularly used to spotting small and large changes between vairables. 
They need to be sometimes be plotted with pivot table

```python
sns.heatmap(df, annot=True, fmt='d')
#annt is used to define it we want to see number in color and fmt is used to declare format
```

As we can clearly see there is an error in the data. This is because the month count is in the form of string <br /> 
Therefore we need to sometimes <b> format our data in a form of pivot </b>

```python
df_pivot = df.pivot("month", "year", "births")
df_pivot
```

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.heatmap(df_pivot, annot = True, fmt = 'd')
plt.show()
```

We can see how beautifully seaborn presented us the data in visual forms using colors. We can easily look at small and large number. For eg- We can say that <b> March 2003, September 2004, December 2004 has small number of births </b> and on the other hand <b> April 2006, January 2006 showed large number of births </b>

### We can even change the color of heatmap using <code>cmap</code>:

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.heatmap(df_pivot, annot = True, fmt = 'd', cmap="RdBu")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.heatmap(df_pivot, annot = True, fmt = 'd', cmap="Blues")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.heatmap(df_pivot, annot = True, fmt = 'd', cmap="Greens")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.heatmap(df_pivot, annot = True, fmt = 'd', cmap="YlGnBu")
plt.show()
```

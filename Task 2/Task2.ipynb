```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = r'E:\Data Visualisation Python Project\rain.json'
df = pd.read_json(path)
df
```

Plotting data:
> We will be using pyplot for plotting the same data

```python
plt.plot(df["Month"], df["Temperature"] , label = 'Temperature')
plt.xlabel("Months")
plt.ylabel("Temperature")
plt.show()
```

We can see the x labels are not clear, so we can rotate then:

```python
plt.plot(df["Month"], df["Temperature"] , label = 'Temperature')
plt.xlabel("Months")
plt.ylabel("Temperature")
plt.xticks(rotation = 90)
plt.show()
```

Another method we can do is by inceasing the plot size if we want our labels to be horizontal:

```python
plt.figure(figsize=(15,9)) #1st argument is width and 2nd arguement is height
plt.plot(df["Month"], df["Temperature"] , label = 'Temperature')
plt.xlabel("Months")
plt.ylabel("Temperature")
plt.show()
```

Plotting rainfall in similar way:

```python
plt.figure(figsize=(15,9))
plt.plot(df["Month"], df["Rainfall"] , label = 'Rainfall')
plt.xlabel("Months")
plt.ylabel("Rainfall")
plt.show()
```

By rotationg the labels:

```python
plt.plot(df["Month"], df["Temperature"] , label = 'Rainfall')
plt.xticks(rotation = 90)
plt.xlabel("Months")
plt.ylabel("Rainfall")
plt.show()
```

We can even use both the functions together to make something better:

```python
plt.figure(figsize=(15,8))
plt.plot(df["Month"], df["Rainfall"] , label = 'Rainfall')
plt.xlabel("Months")
plt.ylabel("Rainfall")
plt.xticks(rotation = 90)
plt.show()
```

We can even plot both the plot together to see side by side comparision:

```python
plt.figure(figsize=(15,8))
plt.plot(df["Month"], df["Temperature"],df["Rainfall"] , label = 'Rainfall & Temperature',)
plt.xlabel("Months")
plt.ylabel("Rainfall & Temperature")
plt.xticks(rotation = 90)
plt.legend(["Temperature","Rainfall"])
plt.show()
```

Another way to plot is by using different plt statements:

```python
plt.figure(figsize=(15,8))
plt.plot(df["Month"], df["Rainfall"], label = "Rainfall" )
plt.plot(df["Month"], df["Temperature"], label = "Temperature" )
plt.legend()
plt.show()
```

<b> While plotting using, we need to give arguement to legend in order from top to bottom when we are drawing using same statement. In case of same different statement, legend identifies their labels and name them accordingly </b>

## Plotting using seaborn:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

We will be working on larger dataset for this task:

```python
path = r"E:\Data Visualisation Python Project\tempYearly.csv"
df = pd.read_csv(path)
df
```

<b> In seaborn we are supposed to define x axis and y axis data unlike in matplotlib.pylot </b>

```python
sns.scatterplot(x = "Year", y = "Temperature", data = df)
```

To change the plot size, there is a slightly different approach in seaborn:

```python
sns.set(rc={'figure.figsize':(15,9)})
sns.scatterplot(x = "Year", y = "Temperature", data = df)
plt.show()
```

We can see a correlation between these two variables but how is this correlation? Strong or Weak? Positive or Negative? To answer this we use <code>regplot</code> used to plot regression

```python
sns.set(rc={'figure.figsize':(15,9)})
sns.regplot(x = "Year", y = "Temperature", data = df)
plt.show()
```

Hence we can see that there is a clear Weak Positive relation between these to parameter.

Let's tyr regression between rainfall and temperature to see that if there are any relationship between them too:

```python
sns.set(rc={'figure.figsize':(15,9)})
sns.regplot(x = "Rainfall", y = "Temperature", data = df)
```

Here we get a steep line between the two parameters, therefore we can safely end up saying that there is a Positive Moderate Correlation between these two parameters, i.e. Temperature and Rainfall

# Plotting using seaborn:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = r"E:\Data Visualisation Python Project\tempYearly.csv"
df = pd.read_csv(path)
df
```

We will be using Joinplot for this task. <br />
Joinplots are widely used to draw many plots at the on the same graph and visualise them.

```python
sns.jointplot(x = "Rainfall", y = "Temperature", data = df)
plt.show()
```

Here we can see many things at the same time from the same visualisation. Like what was the temperature when maximun rainfall happpened or what was the rainfall at the time of maximum temperature.

We can even plot different kind of plot using jointplots:

```python
sns.jointplot(x = "Rainfall", y = "Temperature", data = df, kind = "reg")
plt.show()
```

Here when we choose the kind to regression, we can see a regression line that shows positive relation and alont with that distribution of total rainfall and temperature in form of histogram.
